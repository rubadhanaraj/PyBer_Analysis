```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Add the figure size
plt.figure(figsize=(10,5))

# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Month")
plt.ylabel("Fare($)")

#Set the x limit between Jan through June
plt.xlim(0,5)

# Set the y limit between 0 and 45.
plt.ylim(0,45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add The legend
plt.legend()

# Add a grid.
plt.grid()
```

```python
# Change the line color to green, marker to a diamond,the line width to 2 points.
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(x_axis, y_axis, color="green", marker="D", linewidth=2, label='Boston')

# Add a legend for the city of Boston.
ax.legend()

# Add a title and axes labels.
ax.set_xlabel("Month")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")

# Add a y-axis limit.
ax.set_ylim(0, 45)

# Add grid lines.
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color="tomato", label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
plt.figure(figsize=(20,10))
plt.barh(x_axis, y_axis, color="Magenta", label='Boston')
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color ="k", label='New York')
ax.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color ="cyan", label='Chicago')
ax.legend()
ax.invert_yaxis()
# Add a title and axes labels.
ax.set_ylabel("Month")
ax.set_xlabel("Fare($)")
ax.set_title("PyBer Fare by Month")
```

```python
# Create scatter plot
plt.plot(x_axis,y_axis,'o', c="tomato")
```

```python
# Create scatter plot
plt.scatter(y_axis,x_axis, marker='x', s=y_axis, c='purple', label ='Chicago')
plt.legend()
plt.title("PyBer Fare by Month")
plt.ylabel("Month")
plt.xlabel("Fare($)")
plt.gca().invert_yaxis()
plt.xlim(0,45)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
# Create scatter plot using object oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create scatter plot using object oriented interface.
# Change the color of the markers to sky blue.
# Change the size of the markers to 5 times each data point.
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], c="skyblue",alpha = 0.2, edgecolor = 'black', linewidth=2, label = 'Boston')
ax.legend()
# Add a title and axes labels.
ax.set_xlabel("Month")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
# Invert y axis
ax.invert_yaxis()
# Add a x-axis limit.
ax.set_xlim(0, 45)
```

```python
# Create a pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode = explode_values, labels=x_axis, autopct = '%.1f%%',colors=colors)
plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface
# Add a percentage to one decimal place to each wedge of the pie.
# Increase the figure size to 8x8.
# Explode the two highest percentage months.
# Add a shadow.
# Add a start angle so that January is at the top.
# Reverse the order so that the month order is in a clockwise direction.
# Add new colors of your own choosing or use the colors from the previous pie chart.
fig, ax = plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0.2, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,labels=x_axis,colors=colors,autopct = '%.1f%%', explode=explode_values,shadow=True, startangle=90,counterclock=False)
plt.show()
```
